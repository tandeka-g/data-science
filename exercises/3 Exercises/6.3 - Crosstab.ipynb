# Cross tabulation

The tips dataset contains data one waiter recorded about each tip he received over a period of a few months working in a restaurant. More info about each field can be found on [https://vincentarelbundock.github.io/Rdatasets/doc/reshape2/tips.html](https://vincentarelbundock.github.io/Rdatasets/doc/reshape2/tips.html).

```python
import pandas as pd 

# Import data
df = pd.read_csv("files/tips.csv")
df.head()
```

Put the days in chronological order!

```python
# --- Task 1: Order the 'day' column ---
# Define the custom order explicitly using CategoricalDtype
day_order = pd.CategoricalDtype(categories=['Thur', 'Fri', 'Sat', 'Sun'], ordered=True)
df['day'] = df['day'].astype(day_order)

# Verify the order
print(df['day'].cat.categories)


```

List how many tips the waiter received per day during lunch or dinner.

```python
# --- Task 2: Crosstab of Day vs Time ---
# Using pd.crosstab
day_time_ct = pd.crosstab(df['day'], df['time'])

# Sort the index to ensure chronological order (Thu, Fri, Sat, Sun)
# Since we set the dtype to ordered, sorting should respect that, but let's be explicit
day_time_ct = day_time_ct.sort_index()
print(day_time_ct)

```

You can calculate the row and column totals by using the margins attribute.  

```python
# --- Task 3: Crosstab with Margins ---
day_time_margin = pd.crosstab(df['day'], df['time'], margins=True, margins_name='Total')
print(day_time_margin.sort_index())

```

Next, create a crosstab displaying the number of tips received per gender and per smoker type (rows) per time and per day (columns)

```python
# --- Task 4: Complex Crosstab (Sex/Smoker vs Time/Day) ---
# Using pd.crosstab with multiple indices and columns
complex_ct = pd.crosstab(
    index=[df['sex'], df['smoker']], 
    columns=[df['time'], df['day']]
)
print(complex_ct)

```

Besides getting counts (and percentages) by categories, we can also see aggregations of numeric variables by the categories. For instance, show the average tip amount by day and time.

```python
# --- Task 5: Aggregation with Pivot Table (Alternative to crosstab) ---
# Instead of crosstab, let's use pivot_table to show average tip amount
# This demonstrates an alternative way to get similar aggregated views
tip_avg_pivot = df.pivot_table(
    values='tip', 
    index='day', 
    columns='time', 
    aggfunc=['mean', 'median'], 
    fill_value=0 # Fill NaNs with 0 for cleaner display
)
print(tip_avg_pivot)
```
