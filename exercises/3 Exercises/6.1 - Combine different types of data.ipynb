Create a series of the dictionary and save it as "country_2022_population"

````python
#  country_2022_population
country_2022_population = pd.Series({"Kabul": 41128771, "Tirana": 2842321})
print("country_2022_population")
print(country_2022_population, "\n")


````

Create another series using the following info and save it as "country_rank".

```
rank = [36, 138, 34] and index=["Kabul", "Tirana", "Algiers"]
```

````python

#  country_rank
country_rank = pd.Series([36, 138, 34], index=["Kabul", "Tirana", "Algiers"])
print("country_rank")
print(country_rank, "\n")
````

Make a dataframe combining "country_rank" and "country_2022_population":

```
{"Rank": country_rank,"2022 population": country_2022_population}
```

````python
# dataframe country_data
country_data = pd.DataFrame({"Rank": country_rank, "2022 population": country_2022_population})
print("country_data")
print(country_data, "\n")

````

Print the names of the columns.

````python
#  columns names
print("columns:", country_data.columns.tolist(), "\n")

````

Make a new dataframe and save it as "further_country_data" using the following info:

```
series={"Rank": [213, 203], "2022 population":[44273, 79824]}
index=["Pago Pago", "Andorra la Vella"]
```

````python
# further_country_data
further_country_data = pd.DataFrame(
    {"Rank": [213, 203], "2022 population": [44273, 79824]},
    index=["Pago Pago", "Andorra la Vella"]
)
print("further_country_data")
print(further_country_data, "\n")

````

Concat "country_data" and "further_country_data" as a new dataframe and save it as "all_country_data".

````python
all_country_data = pd.concat([country_data, further_country_data], sort=False)
all_country_data
````

Use the following "capital_countries" dataframe.

Concat "capital_countries" and "all_country_data" as a new dataframe and save it as capitals.

Try sort=False and sort=True, which one is the default?

````python
capital_countries = pd.DataFrame({
"Country": ["Afghanistan", "Albania", "Algeria", "American Samoa", "Andorra"],
"Capital": ["Kabul", "Tirana", "Algiers", "Pago Pago", "Andorra la Vella"]},
index=["Kabul", "Tirana", "Algiers", "Pago Pago", "Andorra la Vella"])
````

````python
# concat to all_country_data
all_country_data = pd.concat([country_data, further_country_data], sort=False)
print("all_country_data")
print(all_country_data, "\n")

````

````python
# capital_countries then concat
capital_countries = pd.DataFrame(
    {
        "Country": ["Afghanistan", "Albania", "Algeria", "American Samoa", "Andorra"],
        "Capital": ["Kabul", "Tirana", "Algiers", "Pago Pago", "Andorra la Vella"],
    },
    index=["Kabul", "Tirana", "Algiers", "Pago Pago", "Andorra la Vella"]
)
capitals = pd.concat([capital_countries, all_country_data], axis=1, sort=False)
print("capitals")
print(capitals, "\n")
# default sort argument in concat is sort=None (i.e. False), so this is same as sort=False.


````

Use the following dataframe "countries_2" and merge it with "capitals" using left_on="Country", right_index=True,how="left".

````python
countries_2 = pd.DataFrame({
"2020 Population": [33428485, 15585, 92664],
"continent": ["Africa", "North America", "North America"]
}, index= ["Afghanistan", "Algeria", "American Samoa"])
````

````python
# merge countries_2 with capitals
countries_2 = pd.DataFrame(
    {
        "2020 Population": [33428485, 15585, 92664],
        "continent": ["Asia", "Africa", "North America"],
    },
    index=["Afghanistan", "Algeria", "American Samoa"]
)
merged_df = countries_2.merge(capitals, left_index=True, right_on="Country", how="left")
print("merged_df")
print(merged_df, "\n")
````

Sort the values of the merged_dataframe by "Rank" and set "ascending" to "False".

````python


#  sort by Rank descending
if "Rank" in merged_df.columns:
    merged_rank_desc = merged_df.sort_values("Rank", ascending=False)
    print("merged_rank_desc")
    print(merged_rank_desc)
else:
    print("Rank not found (maybe missing values), skipping sort.")

````
