# The sales dataset

Sales.csv can be found in the files folder. The file contains 100 sales records. Each record contains 6 fields:
* Quantity
* InvoiceDate
* UnitPrice
* CustomerID
* CountryCode

**Tasks**  
Analyze the dataset using only NumPy. 

**Step 1**   
Load the dataset using np.loadtxt() or np.genfromtxt().

```python
import numpy as np

# Step 1: Load Data
# Using genfromtxt to handle potential mixed types if needed, though sales.csv seems numeric-friendly
sales_db = np.genfromtxt('files/sales.csv', delimiter=',', skip_header=1)
print(f"Dataset dimensions: {sales_db.shape}")
```

**Step 2**   
Calculate basic statistics for the field UnitPrice:   
These basic statistics include min, max, mean, median and standard deviation.

```python
# Step 2: UnitPrice Statistics
# Column index 2 is UnitPrice
prices = sales_db[:, 2]
stats_report = {
    "Mean": np.mean(prices),
    "Median": np.median(prices),
    "Std Dev": np.std(prices)
}
for k, v in stats_report.items():
    print(f"{k}: {v:.2f}")

```

**Step 3**   
Revenue analysis:
* Calculate the overall total revenue
* Calculate the total revenue per country

```python
# Step 3: Revenue Calculations
# Revenue = Qty (col 0) * Price (col 2)
transaction_revenue = sales_db[:, 0] * sales_db[:, 2]
grand_total = np.sum(transaction_revenue)

print(f"\nGrand Total Revenue: ${grand_total:,.2f}")

# Revenue by Country (Col 4)
country_ids = sales_db[:, 4]
unique_countries = np.unique(country_ids)

print("\nRevenue Breakdown by Country:")
for c_id in unique_countries:
    # Boolean masking to sum revenue for specific country
    mask = country_ids == c_id
    rev = np.sum(transaction_revenue[mask])
    print(f"Country {int(c_id)}: ${rev:,.2f}")

```

**Step 4**   
Customer insights:
* Identify the customer with the highest total spend.
* Identify the customer with the most purchases.

```python
# Step 4: Customer Insights
cust_ids = sales_db[:, 3]
unique_custs = np.unique(cust_ids)

# Highest Spender
spend_map = [(cid, np.sum(transaction_revenue[cust_ids == cid])) for cid in unique_custs]
spend_arr = np.array(spend_map)
top_spender_idx = np.argmax(spend_arr[:, 1])
top_spender_id = int(spend_arr[top_spender_idx, 0])

# Most Active Buyer (most transactions)
trans_counts = [(cid, np.sum(cust_ids == cid)) for cid in unique_custs]
trans_arr = np.array(trans_counts)
most_freq_idx = np.argmax(trans_arr[:, 1])
most_freq_id = int(trans_arr[most_freq_idx, 0])
most_freq_count = int(trans_arr[most_freq_idx, 1])

print(f"Highest Spender: Customer {top_spender_id} (${spend_arr[top_spender_idx, 1]:.2f})")
print(f"Most Active Buyer: Customer {most_freq_id} ({most_freq_count} transactions)")
```

**Step 5**   
Time analysis:
* Convert timestamps to NumPy datetime64.
* Group by month and calculate monthly revenue.

```python
# Step 5: Time Analysis
# Convert timestamps (col 1) to datetime64
dates = sales_db[:, 1].astype('datetime64[D]') # Day precision first
months = dates.astype('datetime64[M]') # Cast to Month
unique_months = np.unique(months)

print("\nMonthly Revenue Trend:")
for m in unique_months:
    m_mask = months == m
    m_rev = np.sum(transaction_revenue[m_mask])
    # Format date nicely
    print(f"{str(m)}: ${m_rev:,.2f}")

```
