### Exercise 1
Nicholas and Jordan went on holidays with their families. They decided to monitor the kilometers they traveled by keeping track of the time and the distance they were on the road. The boys collected the following data:  

Nicholas:  

Time (hr)| 1 | 2 | 3 | 4 | 5 | 6
--------|---|---|---|---|---|---
 **Distance (kilometers)**| 60 | 110 | 175 | 235 | 280 | 320


Jordan:
Time (hr)| 1 | 2 | 3 | 4 | 5 | 6
--------|---|---|---|---|---|---
 **Distance (kilometers)**| 50 | 90 | 125 | 125 | 165 | 210

* Draw a graph to show the trip of each boy.  
* What conclusions could you draw by looking at the graphs?

![](files/2023-11-16-08-42-06.png)

````python
import numpy as np 

nicholas = np.array([0,60,110,175,235,280,320])
jordan = np.array([0,50,90,125,125,165,210])
````

````python
import numpy as np
import matplotlib.pyplot as plt

# ### Exercise 1: Holiday Trip
nicholas_dist = np.array([0, 60, 110, 175, 235, 280, 320])
jordan_dist = np.array([0, 50, 90, 125, 125, 165, 210])
time_axis = np.arange(len(nicholas_dist))

plt.figure(figsize=(10, 6))
plt.plot(time_axis, nicholas_dist, marker='o', label='Nicholas', linewidth=2)
plt.plot(time_axis, jordan_dist, marker='s', label='Jordan', linewidth=2)
plt.xlabel('Time (hours)')
plt.ylabel('Distance (km)')
plt.title('Holiday Trip Distance Comparison')
plt.xticks(time_axis)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

# Conclusion update
print("Observation: Nicholas maintained a consistent velocity throughout the trip. Jordan, however, plateaued between hour 3 and 4, suggesting a rest stop, before resuming travel at a slower average pace than Nicholas.")



````

### Exercise 4

The following table consists of the marks achieved by 9 students on chemistry and math tests:


Student | A | B | C | D | E | F | G | H | I
--------|---|---|---|---|---|---|---|---|--
 **Chemistry marks** | 49 | 46 | 35 | 58 | 51 | 56 | 46 | 53 | 39
 **Math marks** | 29 | 23 | 10 | 40 | 38 | 36 | 32 | 28| 18 | 

The question is whether students with high marks in maths are also good at chemistry.  Choose a graph type suitable for investigating whether or not there a relationship or connection exists between 2 sets of data.



````python
import numpy as np
chemistry_marks = np.array([49, 46,35,58,51,56, 46, 35, 39])
math_marks = np.array([29, 23,10,40,38,36, 32, 28, 18])
````

````python
### Exercise 4: Chemistry vs Math
chem_scores = np.array([49, 46, 35, 58, 51, 56, 46, 53, 39])
math_scores = np.array([29, 23, 10, 40, 38, 36, 32, 28, 18])

plt.figure(figsize=(8, 6))
plt.scatter(math_scores, chem_scores, color='purple', s=80, alpha=0.7)
plt.xlabel('Math Scores')
plt.ylabel('Chemistry Scores')
plt.title('Chemistry vs Math Performance')
plt.grid(True, linestyle='--', alpha=0.4)

# Optional trend line
m, b = np.polyfit(math_scores, chem_scores, 1)
plt.plot(math_scores, m*math_scores + b, color='black', linestyle='--', label=f'Trend: y={m:.2f}x+{b:.1f}')
plt.legend()
plt.show()

print("Conclusion: There is a visible positive trend; students scoring higher in Math tend to score higher in Chemistry. The points cluster along an imaginary diagonal line, indicating a moderate-to-strong positive correlation between aptitude in these two subjects.")
````

### Exercise 5

The Red Cross Blood Donor Clinic had a very successful morning collecting blood donations. Within 3 hours, 25 people had made donations, and the following is a table showing the blood types of the donations:

Blood Type | A | B | O | AB
-----------|---|---|---|---
 **Number of donors** | 9 | 5 | 7 | 4

Plot the donations so that it is clear what percentage of each blood type was collected that morning.


````python
import numpy as np  

blood_types = np.array(['A','B','O','AB'])
number_of_donors = np.array ([9,5,7,4])
````

````python
# ### Exercise 5: Blood Donations
blood_types = ['A', 'B', 'O', 'AB']
donor_counts = [9, 5, 7, 4]

plt.figure(figsize=(7, 7))
# Pie chart customization: explode one slice, different colors, shadow
explode_vals = (0.1, 0, 0, 0)  # Explode 'A' slightly
colors_custom = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.pie(donor_counts, labels=blood_types, autopct='%1.1f%%', startangle=90, 
        explode=explode_vals, colors=colors_custom, shadow=True, textprops={'fontsize': 12})
plt.title('Blood Type Distribution Among Donors', fontsize=14, fontweight='bold', pad=20)
plt.axis('equal')  
plt.show()

````
