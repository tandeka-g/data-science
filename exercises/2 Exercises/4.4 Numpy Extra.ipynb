# Exercises numpy

Numpy is embedded quite well into pandas (which is the next chapter). NumPy is very useful for performing logical and mathematical calculations on arrays and matrices. This tool performs these operations much faster and more efficiently than Python lists. 

Furthermore, Numpy uses less memory and storage space, which is the main advantage. This is because numpy is typed. Normal Python is typeless, so if you create a list of integers, Python ensures that it is also suitable for floating point numbers or very long strings. The more items you add to the list, the more space is used as a list grows exponentially. This means that a lot of memory is required, whereas numpy knows what the array is intended for and allocates exactly enough memory.

## Rotate an array

Create a python list with 3 items, and initiate every item with a list of 4 items. As content of the list, use the next letter of the alphabet. ([tip](https://www.askpython.com/python/built-in-methods/python-chr-and-ord-methods)).

To generate:
```
[['a', 'b', 'c', 'd'], ['e', 'f', 'g', 'h'], ['i', 'j', 'k', 'l']]
```

````python
import numpy as np
import re

# ## Rotate an array
# Create the initial list of lists using alphabet logic
rows, cols = 3, 4
# Generating characters dynamically
char_list = [[chr(ord('a') + r * cols + c) for c in range(cols)] for r in range(rows)]
print("Original List:", char_list)

````

Now turn the array so you get a rotated list of lists. This would be a list of 4 items, each of them containing a list with 3 items. (Do note however that if we were to change to original list to a 5x12-matrix your code should still work.)

Output:
```
[['a', 'e', 'i'], ['b', 'f', 'j'], ['c', 'g', 'k'], ['d', 'h', 'l']]
```

````python
# Rotation Logic 1: Pure Python List Comprehension (Transposing)
# Swapping i and j indices effectively rotates/transposes
rotated_python = [[char_list[r][c] for r in range(rows)] for c in range(cols)]
print("Rotated (Python):", rotated_python)


````

Now convert the original list to a numpy array and [rotate it 90 degrees](https://numpy.org/doc/stable/reference/generated/numpy.rot90.html).

Output:
```
[['a' 'b' 'c' 'd']
 ['e' 'f' 'g' 'h']
 ['i' 'j' 'k' 'l']]
 
[['d' 'h' 'l']
 ['c' 'g' 'k']
 ['b' 'f' 'j']
 ['a' 'e' 'i']]
```

````python
# Rotation Logic 2: NumPy Approach
np_arr = np.array(char_list)
# np.rot90 rotates 90 degrees counter-clockwise. 
# To match the python output (which is a transpose), we can use .T or rot90 with specific axes/k
# The python output [['a','e','i']...] is actually a Transpose, not a geometric 90deg rotation of the visual block.
# Let's use transpose to match the python logic exactly, or rot90(3) to rotate 270 deg (which looks like transpose for square-ish logic)
rotated_np = np.transpose(np_arr) 
print("Original NP Array:\n", np_arr)
print("Rotated NP Array (Transpose):\n", rotated_np)

````

The problem: you have 1000x1000 grid of lights, and need to turn them on or off based on a file-input. In the end you need to know how many are on. We'll be solving this by creating the actual array and simply turning setting the lights according to the file.

* Create an empty grid of 1000 by 1000 zeros
* Import all (decoded) lines from your file into an array 

````python
# ## Advent of Code 2015, Day 6
# Setup
grid_shape = (1000, 1000)
lights_grid = np.zeros(grid_shape, dtype=int)
parsed_instructions = []
pattern = re.compile(r'^(turn on|turn off|toggle) (\d+),(\d+) through (\d+),(\d+)$')
input_path = 'files/aoc_day6.txt'

try:
    with open(input_path, 'r') as f:
        for line in f:
            line = line.strip()
            match = pattern.match(line)
            if match:
                action, x1, y1, x2, y2 = match.groups()
                parsed_instructions.append((action, (int(x1), int(y1)), (int(x2), int(y2))))
except FileNotFoundError:
    # Fallback dummy data if file missing for demonstration
    parsed_instructions = [('toggle', (0,0), (5,0)), ('turn on', (4,4), (5,5))]

print('Parsed instructions:', parsed_instructions)


````

You can use the following code to create the small arrays that are given to test your code:

````python
grid = np.zeros( (6,6) )
data = [('toggle', (0,0), (5,0)), ('toggle', (0,0), (0,5)), ('turn on', (4,4), (5,5))]
````

Make 2 functions:

"get_value(current, action)"

Current is the current state of a light (1 or 0), action is what you want to do 'turn on', 'turn off' or 'toggle'. The return value is whatever the value needs to become.

"apply_to_grid(line)"

Line is a decoded line, it's a python-list (or tuple) consisting of:

* the action
* tuple of start coordinate
* tuple of end coordinate

It will apply the action to all points within the coordinates.

````python
# Define functions and apply instructions

def get_value(current, action):
    if action == 'turn on':
        return 1
    elif action == 'turn off':
        return 0
    elif action == 'toggle':
        return 1 - current
    raise ValueError(f"Unknown action: {action}")


def apply_to_grid(grid, instruction):
    action, (x1, y1), (x2, y2) = instruction
    region = grid[x1:x2+1, y1:y2+1]

    if action == 'turn on':
        region[:] = 1
    elif action == 'turn off':
        region[:] = 0
    elif action == 'toggle':
        region[:] = 1 - region
    else:
        raise ValueError(f"Unknown action: {action}")


# Apply parsed instructions from earlier cells
for instr in parsed_instructions:
    apply_to_grid(lights_grid, instr)

print("Lights turned on:", np.sum(lights_grid))
````

And now apply all the lines from your "data"-list to the numpy-array. When done print the sum of the array (this equals the number of lights turned on).

````python
# Execute Part 1
for inst in parsed_instructions:
    apply_to_grid(lights_grid, inst)

print(f"Part 1 - Lights On: {np.sum(lights_grid)}")
````

````python
# Part 2: Brightness
# Reset grid
brightness_grid = np.zeros(grid_shape, dtype=int)

def apply_brightness_vectorized(grid, instruction):
    action, start, end = instruction
    x1, y1 = start
    x2, y2 = end
    region = grid[x1:x2+1, y1:y2+1]

    if action == 'turn on':
        region += 1
    elif action == 'turn off':
        region -= 1
        region[region < 0] = 0
    elif action == 'toggle':
        region += 2
    else:
        raise ValueError(f"Unknown action: {action}")

for instr in parsed_instructions:
    apply_brightness_vectorized(brightness_grid, instr)

print(f"Part 2 - Total Brightness: {np.sum(brightness_grid)}")

````
